```python
import time

TAX_PERCENTAGE = 0.05
def unique_items_in_store(store_dict):
    # get the list of keys in dict and find the length of that list
    return len(store_dict.keys())

def total_num_of_transactions(transaction_data):
    # get the length of the transaction_data array
    return len(transaction_data)

def add_item_to_store(store_dict, item, value):
    # add new item to dictionary
    store_dict[item] = value

def delete_item_from_store(store_dict, item):
    # check if item is present. If yes, delete it
    if item in store_dict:
        del store_dict[item]

def get_cost_of_item(store_dict, item):
    # Check if item is in dict. If not return None.
    if item in store_dict:
        return store_dict[item]
    
    return None

def get_nth_transaction(transaction_data, index):
    # Check if index is in range of transaction data
    if index < len(transaction_data):
        return transaction_data[index - 1]
    return None

def get_num_items_in_transaction(transaction_data, index):
    # Get the list from transaction_data
    transaction = get_nth_transaction(transaction_data, index)
    # Length of the transaction
    if transaction:
        return len(transaction)
    return 0
        
def add_transaction_to_data(frequency_dict, transaction_data, item_list):
    if len(item_list):
        transaction_data.append(item_list)
        update_counter(frequency_dict, item_list)
        
def total_cost_of_nth_transaction(transaction_data, store_dict, index):
    transaction = get_nth_transaction(transaction_data, index)
    total_amount = 0
    for item in transaction:
        total_amount += store_dict[item]
    return total_amount

def find_total(transaction_data, store_dict, index):
    item_costs = total_cost_of_nth_transaction(transaction_data, store_dict, index)
    taxes = TAX_PERCENTAGE * item_costs
    return item_costs + taxes

def generate_frequency_dict(frequency_dict, transaction_data):
    for transaction in transaction_data:
        update_counter(frequency_dict, transaction)

def update_counter(frequency_dict, transaction):
    for item in transaction:
        if item in frequency_dict:
            frequency_dict[item] += 1
        else:
            frequency_dict[item] = 1

def top_selling_item(input_dict):
    return max(zip(input_dict.values(), input_dict.keys()))[1]
    

def get_list_from_dict(frequency_dict):
    return list(zip(frequency_dict.keys(), frequency_dict.values()))

def merge(frequency_list, start_index, mid, end_index):
    # First half length of the given list
    first_half_length = mid - start_index + 1
    # Second half length of the given list
    second_half_length = end_index - mid
 
    # Create a left side array 
    left_array = [0] * (first_half_length)
    # Create right side array
    right_array = [0] * (second_half_length)
 
    # Copy the content of first half to left array 
    for i in range(0, first_half_length):
        left_array[i] = frequency_list[start_index + i]
 
    # Copy the content of second half to right array
    for j in range(0, second_half_length):
        right_array[j] = frequency_list[mid + 1 + j]
        
    i = 0
    j = 0
    k = start_index

    # Merge the left and right arrays
    while i < first_half_length and j < second_half_length:
        if left_array[i][1] <= right_array[j][1]:
            frequency_list[k] = left_array[i]
            i += 1
        else:
            frequency_list[k] = right_array[j]
            j += 1
        k += 1
    
    # Copy remaining items of left array
    while i < first_half_length:
        frequency_list[k] = left_array[i]
        i += 1
        k += 1
    
    # Copy remaining items of right array
    while j < second_half_length:
        frequency_list[k] = right_array[j]
        j += 1
        k += 1
 
 
def mergeSort(frequency_list, start_index, end_index):
    if start_index < end_index:
 
        mid = start_index+(end_index-start_index)//2
 
        # Sort first and second halves
        mergeSort(frequency_list, start_index, mid)
        mergeSort(frequency_list, mid+1, end_index)
        # merge sorted arrays
        merge(frequency_list, start_index, mid, end_index)

def insertion_sort(frequency_list):
    for i in range(1, len(frequency_list)):
        key = frequency_list[i][1]
        tuple_value = frequency_list[i]
        j = i-1
        while j >=0 and key < frequency_list[j][1] :
                frequency_list[j+1] = frequency_list[j]
                j -= 1
        frequency_list[j+1] = tuple_value
        
def selection_sort(frequency_list):
    for i in range(len(frequency_list)):
        index = i
        for j in range(i+1, len(frequency_list)):
            if frequency_list[index][1] > frequency_list[j][1]:
                index = j
              
        frequency_list[i], frequency_list[index] = frequency_list[index], frequency_list[i]

def milliseconds_taken(start_time, end_time):
    return round((end_time - start_time) * 1000, 3) 

def top_three_selling_items(frequency_dict):
    frequency_list = get_list_from_dict(frequency_dict)
    mergeSort(frequency_list, 0, len(frequency_list)-1)
    return([frequency_list[-1][0], frequency_list[-2][0], frequency_list[-3][0]])

def merge_sort_frequency_list(frequency_list):
    frequency_list_merge_sort = copy.deepcopy(frequency_list)
    start_time = time.time()
    mergeSort(frequency_list_merge_sort, 0, len(frequency_list_merge_sort)-1)
    end_time = time.time()
    # Record sorting time of merge_sort
    return milliseconds_taken(start_time, end_time)

def insertion_sort_frequency_list(frequency_list):
    frequency_list_insertion_sort = copy.deepcopy(frequency_list)
    start_time = time.time()
    insertion_sort(frequency_list_insertion_sort)
    end_time = time.time()
    # Record sorting time of merge_sort
    return milliseconds_taken(start_time, end_time)

def selection_sort_frequency_list(frequency_list):
    frequency_list_selection_sort = copy.deepcopy(frequency_list)
    start_time = time.time()
    insertion_sort(frequency_list_selection_sort)
    end_time = time.time()
    # Record sorting time of merge_sort
    return milliseconds_taken(start_time, end_time)

# 1.1. Find how many unique items are present in the data? => 47
print("Unique number of items in store: {}".format(unique_items_in_store(item_dict)))
# 1.2. Find the total number of transactions? => 20
print("Total number of transactions {}".format(total_num_of_transactions(transaction_data)))
# 1.3 and 1.4 Add items, try adding the item “banana” in stores. The cost of this item is $5.
# Add the item “banana” to the Cost dictionary. 
add_item_to_store(item_dict, 'banana', 5)
# Items in store after adding Banana => 48
print("Items in store now: {}".format(unique_items_in_store(item_dict)))
# 1.5 Remove an item: for example: if the customers at Walmart do not seem to be purchasing “banana,” 
# and the store decides to discontinue this item. The function to should be able to remove the item 
# “banana” from the Cost dictionary
delete_item_from_store(item_dict, 'banana')
# Items in store after deleting Banana => 47
print("Items in store: {}".format(unique_items_in_store(item_dict)))


# 2.1. Find the price of salmon from the cost dictionary => $3
print("Price of salmon : ${}".format(get_cost_of_item(item_dict, "salmon")))
# 2.2 Try finding the ninth transaction from the data => ['frozen vegetables', 'spaghetti', 'green tea']
print("Ninth transaction : {}".format(get_nth_transaction(transaction_data, 9)))
# 2.3 How many items are present in the ninth transaction? => 3
print("Number of items in ninth transaction {}".format(get_num_items_in_transaction(transaction_data, 9)))
# 2.4 Find the total cost of the ninth transaction from the array of transactions. => $16
print("Cost of transaction without tax {}".format(total_cost_of_nth_transaction(transaction_data, item_dict, 9)))
# 2.5 and 2.6 As per government guidelines, a mandatory 5% tax is applied to the total cost of a transaction.
# Find the final price of the second transaction after applicable taxes have been implemented. => $ 23.1
print("Total amount including tax {}".format(find_total(transaction_data, item_dict, 2)))


# 3.1 What is the data structure that can be used to count the frequency of items purchased (from the transaction data)?
# Ans. Dictionary data structure will be used, where key is the item name and value is the frequency of purchase.
      
# 3.2 Create a frequency counter to count the number of items:
frequency_dict = {}
generate_frequency_dict(frequency_dict, transaction_data)
print(frequency_dict)
      
# 3.3 and 3.4 A new transaction has just been made. This transaction includes “honey,” “pet food,” and “shrimp.”
#      Update the frequency counter to reflect this new transaction.
#      Create a function named update_counter( ) to automatically update the frequency counter in case any
#      new transaction is made.
add_transaction_to_data(frequency_dict, transaction_data, ['honey', 'pet food', 'shrimp'])
      
      
# 4 Sort the frequency dict with merge, insertion and selection sort
      
# convert frequency dict into a frequency list with each index value tuple of (key, value)
frequency_list = get_list_from_dict(frequency_dict)
import copy
sorting_time = {}
sorting_time['merge_sort'] = merge_sort_frequency_list(frequency_list)
sorting_time['insertion_sort'] = insertion_sort_frequency_list(frequency_list)
sorting_time['selection_sort'] = selection_sort_frequency_list(frequency_list)

# 4 => Top three selling items => [mineral_water, turkey, eggs] or [mineral_water, turkey, honey] as eggs and honey are sold
# equal number of times
print("Top selling three items {}".format(top_three_selling_items(frequency_dict)))

# 4.1 => Top selling item => Mineral Water
print("Top selling item : {}".format(top_selling_item(frequency_dict)))

# Algorithm  Time
# Merge  => 0.0009987354278564453
# Insertion => 0.0010025501251220703
# Selection  => 0.0011025501251220703
# Sort the algorithms as per the time they took
sort_time_list = get_list_from_dict(sorting_time)
selection_sort(sort_time_list)
print(dict(sort_time_list))
```
